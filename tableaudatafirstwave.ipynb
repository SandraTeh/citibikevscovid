```python
#get the file
mac_df = "data/202003-citibike-tripdata.csv"
apr_df = "data/202004-citibike-tripdata.csv"
may_df = "data/202005-citibike-tripdata.csv"
jun_df = "data/202006-citibike-tripdata.csv"
```

```python
#read the data file
df_mac = pd.read_csv(mac_df)
df_apr = pd.read_csv(apr_df)
df_may = pd.read_csv(may_df)
df_jun = pd.read_csv(jun_df)
```

```python
#combine citibike dataframe
frames = [df_mac, df_apr, df_may, df_jun]
```

```python
fcitibike = pd.concat(frames)
```

```python
fcitibike.head()
```

```python
fcitibike["starttime"] = pd.to_datetime(fcitibike["starttime"])
```

```python
fcitibike["date"] = fcitibike["starttime"].dt.date
```

```python
fcitibike_by_date = fcitibike.groupby([fcitibike['date'],fcitibike['start station latitude'],fcitibike['start station longitude']]).starttime.count()
```

```python
fcitibike_by_date = pd.DataFrame(fcitibike_by_date).reset_index()
```

```python
fcitibike_by_date = fcitibike_by_date.rename(columns={"date": "date","start station latitude": "lat","start station longitude": "lon","starttime": "no_of_trip"})
```

```python
fcitibike_by_date.head()
```

```python
fcitibike_by_date.to_csv('data\\firstwavetripmap.csv')
```
