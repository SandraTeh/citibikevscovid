```python
#Import Pandas
import pandas as pd

#Import datetime
import datetime
```

```python
#get the file
covidcases_df = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv"
```

```python
#read the data file
df_covidcases = pd.read_csv(covidcases_df)
```

```python
df_covidcases.head()
```

```python
df_covidcases = df_covidcases.loc[df_covidcases["state"] == "New York",:].reset_index(drop=True)
```

```python
df_covidcases.head(30)
```

```python
print(df_covidcases.dtypes)
```

```python
df_covidcases["date"] = pd.to_datetime(df_covidcases["date"])
```

```python
print(df_covidcases.dtypes)
```

```python
start_date = '2020-09-30'
end_date = '2021-01-31'
mask = (df_covidcases["date"] >= start_date) & (df_covidcases["date"] <= end_date)
```

```python
df_covidcases = df_covidcases.loc[mask].sort_values(["county", "date"], ascending = (True, True))
df_covidcases.head(20)
```

```python
df_covidcases["country"] = df_covidcases["county"] + ", " + df_covidcases["state"]
df_covidcases.head(20)
```

```python
df_covidcases["cases_per_day"] = df_covidcases["cases"].diff();
df_covidcases.head()
```

```python
start_date = '2020-10-01'
end_date = '2021-01-31'
mask = (df_covidcases["date"] >= start_date) & (df_covidcases["date"] <= end_date)
```

```python
df_covidcases = df_covidcases.loc[mask].reset_index(drop=True)
df_covidcases.head(20)
```

```python
#get the file
oct_df = "data/202010-citibike-tripdata.csv"
nov_df = "data/202011-citibike-tripdata.csv"
dec_df = "data/202012-citibike-tripdata.csv"
jan_df = "data/202101-citibike-tripdata.csv"
```

```python
#read the data file
df_oct = pd.read_csv(oct_df)
df_nov = pd.read_csv(nov_df)
df_dec = pd.read_csv(dec_df)
df_jan = pd.read_csv(jan_df)
```

```python
#combine citibike dataframe
frames = [df_oct, df_nov, df_dec, df_jan]
```

```python
citibike = pd.concat(frames)
```

```python
citibike.head()
```

```python
citibike["starttime"] = pd.to_datetime(citibike["starttime"])
```

```python
citibike["date"] = citibike["starttime"].dt.date
```

```python
citibike_by_date = citibike.groupby([citibike['date'],citibike['start station latitude'],citibike['start station longitude']]).starttime.count()
```

```python
citibike_by_date = pd.DataFrame(citibike_by_date).reset_index()
```

```python
citibike_by_date = citibike_by_date.rename(columns={"date": "date","start station latitude": "lat","start station longitude": "lon","starttime": "no_of_trip"})
```

```python
citibike_by_date.head()
```

```python
citibike_by_date.to_csv('data\secondwavetripmap.csv')
```
